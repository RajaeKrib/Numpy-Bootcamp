### <a name='1'></a> Exercise 31
Sort the array $ A $ below:
* by row (ascending)
* by column (ascending)

```
A = np.array([[4.99, 3.49, 9.99],
              [1.99, 9.99, 4.99],
              [14.99, 2.39, 7.29]])
```
__Tip:__ Use the _np.sort()_ function.


````
A = np.array([[4.99, 3.49, 9.99],
              [1.99, 9.99, 4.99],
              [14.99, 2.39, 7.29]])

#row ascending 
print(f'Row ascending: \n {np.sort(A)}')

#column ascending
print(f'Column ascending: \n {np.sort(A, axis=0)}')
````

### <a name='3'></a> Exercise 33
The following array $ A $ is given. Replace elements above 10 with a fixed value of 10 (truncation to 10).

```
A = np.array([[4.99, 3.49, 9.99],
              [1.99, 9.99, 14.99],
              [14.99, 2.39, 7.29]])
```
__Tip:__ Use the _np.where()_ function.


````
A = np.array([[4.99, 3.49, 9.99],
              [1.99, 9.99, 14.99],
              [14.99, 2.39, 7.29]])

A[A>10]=10
#or 
np.where(A>10,10,A)
````
